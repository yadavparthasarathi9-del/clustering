#1 . What is unsupervised learning in the context of machine learning?

-> Unsupervised learning is a type of machine learning in which the algorithm learns from unlabeled data, meaning no target or output variable is provided. The main objective of unsupervised learning is to discover hidden patterns, structures, or relationships within the data without any prior knowledge of the outcomes.

In unsupervised learning, the model analyzes the input data and groups similar data points together or reduces the complexity of the data. Common techniques include clustering, dimensionality reduction, and association rule learning.

#2. How does K-Means clustering algorithm work?

-> K-Means is an unsupervised clustering algorithm used to group data into K distinct clusters based on similarity. The algorithm works by minimizing the within-cluster sum of squares (WCSS), which measures the distance between data points and their assigned cluster centroids.

Steps of K-Means Algorithm:

-  the number of clusters (K) to be formed.

- Initialize K centroids randomly from the data points.

- Assignment step: Each data point is assigned to the nearest centroid using a distance measure such as Euclidean distance.

- Update step: New centroids are calculated by taking the mean of all data points assigned to each cluster.

- Repeat steps 3 and 4 until the centroids no longer change or the assignments become stable.

The algorithm converges when there is no significant change in cluster centroids. K-Means is simple, efficient, and widely used in applications like customer segmentation, image compression, and pattern recognition, but it requires the value of K to be predefined and is sensitive to initial centroid selection.

#3. Explain the concept of a dendrogram in hierarchical clustering?

-> A dendrogram is a tree-like diagram used in hierarchical clustering to represent how data points are grouped into clusters based on similarity. Each leaf node represents a data point, and branches show the merging of clusters.

The height of the branches indicates the distance or dissimilarity at which clusters are joined. Clusters merging at lower heights are more similar. By drawing a horizontal cut across the dendrogram, we can determine the number of clusters.

Thus, a dendrogram helps visualize the hierarchical relationship and structure of clusters in the data.

#4. What is the main difference between K-Means and Hierarchical Clustering?

-> K-Means clustering requires the number of clusters (K) to be specified in advance. It forms clusters by iteratively assigning data points to the nearest centroid and updating the centroids until convergence. It is computationally efficient and suitable for large datasets but is sensitive to initial centroid selection.

Hierarchical clustering, on the other hand, does not require the number of clusters beforehand. It builds a hierarchy of clusters either by agglomerative (bottom-up) or divisive (top-down) approaches and represents the results using a dendrogram. It is more interpretable but computationally expensive for large datasets.

In summary, K-Means is faster and requires predefined K, while hierarchical clustering provides a hierarchical structure without needing K in advance.

#5. What are the advantages of DBSCAN over K-Means?

-> The main advantage of DBSCAN :    

Firstly, DBSCAN does not require the number of clusters to be specified in advance, whereas K-Means needs the value of K beforehand. Secondly, DBSCAN can identify clusters of arbitrary shape, while K-Means mainly works well with spherical or circular clusters.

Another major advantage of DBSCAN is its ability to detect noise and outliers. Data points that do not belong to any cluster are labeled as noise, which K-Means cannot handle effectively. DBSCAN is also less sensitive to initial conditions, unlike K-Means which depends on random centroid initialization.

#6.  When would you use Silhouette Score in clustering?

-> The Silhouette Score is used in clustering to evaluate the quality of clusters formed by a clustering algorithm. It measures how well each data point fits within its assigned cluster compared to other clusters.

Silhouette Score is mainly used when there are no true labels available, which is common in unsupervised learning. It helps in selecting the optimal number of clusters, especially for algorithms like K-Means or Hierarchical Clustering. A higher silhouette score indicates that data points are well clustered, with high intra-cluster similarity and low inter-cluster similarity.

It is also useful for comparing different clustering algorithms or distance metrics on the same dataset. The score ranges from –1 to +1, where values close to +1 indicate well-separated clusters.

#7.  What are the limitations of Hierarchical Clustering?

-> Limitations of Hierarchical Clustering

Hierarchical clustering has several limitations despite its usefulness in understanding data structure. One major limitation is its high computational complexity, making it unsuitable for large datasets. As the number of data points increases, the time and memory requirements grow significantly.

Another limitation is that hierarchical clustering is sensitive to noise and outliers, which can distort the clustering structure. Once a merge or split is made, it cannot be undone, meaning early mistakes affect the final clusters. The results also depend heavily on the choice of distance metric and linkage method

#8. Why is feature scaling important in clustering algorithms like K-Means?

-> Feature scaling is important in clustering algorithms like K-Means because the algorithm relies on distance calculations, typically Euclidean distance, to group data points. If features are on different scales, features with larger values will dominate the distance computation and bias the clustering results.

Without feature scaling, variables with higher ranges (e.g., income) can overshadow variables with smaller ranges (e.g., age), even if they are less important. This leads to incorrect or misleading clusters. Feature scaling ensures that all features contribute equally to the distance measurement.

#9 . How does DBSCAN identify noise points?

-> DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points based on data density using two parameters: ε (epsilon) and MinPts.

- ε (epsilon) defines the maximum distance within which points are considered neighbors.

- MinPts is the minimum number of points required to form a dense region.

A data point is classified as a core point if it has at least MinPts points within its ε-neighborhood. A border point lies within the ε-neighborhood of a core point but has fewer than MinPts neighbors. Any point that is neither a core point nor reachable from a core point is labeled as noise (outlier).

#10. Define inertia in the context of K-Means?

-> In K-Means clustering, inertia refers to the within-cluster sum of squared distances (WCSS) between each data point and the centroid of the cluster to which it is assigned. It measures how compact the clusters are.A lower inertia value indicates that data points are closer to their centroids, resulting in tighter and more cohesive clusters.

Inertia is commonly used to evaluate and compare K-Means models for different values of K. It is the basis of the Elbow Method, where inertia is plotted against K to find an optimal number of clusters

#11.  What is the elbow method in K-Means clustering?

-> The Elbow Method is a technique used to determine the optimal number of clusters (K) in K-Means clustering. It is based on analyzing how the inertia changes as the value of K increases.

In this method, K-Means is run for different values of K, and the corresponding inertia values are plotted against K. As K increases, inertia decreases because clusters become smaller and more compact. At a certain point, the rate of decrease in inertia slows down significantly, forming a shape similar to an “elbow” in the graph.

#12.  Describe the concept of "density" in DBSCAN.

-> In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density refers to how closely data points are packed together in a region of the data space. DBSCAN identifies clusters as areas of high point density separated by areas of low density.

Density is defined using two parameters: ε (epsilon), which specifies the neighborhood radius, and MinPts, the minimum number of points required within that radius. If a point has at least MinPts points within its ε-neighborhood, it is considered a core point and lies in a dense region.

#13.  Can hierarchical clustering be used on categorical data?

-> Yes, hierarchical clustering can be applied to categorical data, but it requires a suitable distance or similarity measure instead of standard Euclidean distance, which only works for numerical data.

For categorical data, common measures include:

- Hamming distance – counts the number of mismatches between categorical attributes.

- Matching coefficient – calculates similarity based on matching attribute values.

- Gower distance – handles mixed data types (numerical + categorical).

Once a distance matrix is computed using these measures, hierarchical clustering can proceed with agglomerative or divisive methods, and the results can be visualized with a dendrogram.

#14.  What does a negative Silhouette Score indicate?

-> A negative Silhouette Score indicates that a data point is misclassified or assigned to the wrong cluster. It occurs when the average distance to points in other clusters is smaller than the average distance to points in its own cluster.

In other words, the point is closer to a neighboring cluster than to its assigned cluster, suggesting poor clustering or overlapping clusters.

#15. Explain the term "linkage criteria" in hierarchical clustering?

-> In hierarchical clustering, linkage criteria determine how the distance between two clusters is calculated when merging or splitting them. The choice of linkage affects the shape and structure of the clusters.

Common linkage methods include:

- Single linkage – distance between the closest pair of points from two clusters.

- Complete linkage – distance between the farthest pair of points.

- Average linkage – average distance between all pairs of points across the clusters.

- Ward’s linkage – minimizes the increase in total within-cluster variance after merging.

#16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

-> K-Means clustering assumes that clusters are roughly spherical, of similar size, and have similar densities. When the data has clusters of different sizes or densities, K-Means can perform poorly because:

1. Centroid bias – The algorithm assigns points based on distance to centroids. Large or dense clusters can dominate, causing small or sparse clusters to be misclassified.

2. Inability to detect irregular shapes – K-Means works best for circular clusters; clusters with elongated or irregular shapes may be split incorrectly.

3. Sensitivity to outliers – Sparse clusters or noise can distort centroid positions, further degrading performance.

#17. What are the core parameters in DBSCAN, and how do they influence clustering?

-> DBSCAN relies on two core parameters to form clusters:

1. ε (epsilon) – the neighborhood radius around a point.

- Determines how close points must be to each other to be considered neighbors.

- A small ε may result in many points being classified as noise, while a large ε can merge distinct clusters into one.

2. MinPts – the minimum number of points required within the ε-neighborhood to form a dense region.

- A higher MinPts makes the algorithm stricter, requiring denser clusters, potentially labeling sparse clusters as noise.

- A lower MinPts allows smaller clusters but may merge noise into clusters.

#Influence on Clustering:

- These parameters control cluster size, shape, and density sensitivity.

- Proper tuning is essential: together they determine which points are core points, border points, or noise.

#18. How does K-Means++ improve upon standard K-Means initialization?

-> K-Means++ is an improved version of the standard K-Means algorithm that focuses on better initialization of cluster centroids to avoid poor clustering results.

K-Means++ initialization works as follows:

1. Choose the first centroid randomly from the data points.

2. For each remaining point, calculate its distance squared from the nearest chosen centroid.

3. Select the next centroid probabilistically, giving higher chance to points far from existing centroids.

4. Repeat until K centroids are initialized, then run standard K-Means.

Advantages:

- Leads to better spread of initial centroids

- Reduces the risk of poor local minima

- Improves convergence speed and cluster quality

#19. What is agglomerative clustering?

-> Agglomerative clustering is a type of hierarchical clustering that follows a bottom-up approach. In this method, each data point starts as its own individual cluster, and clusters are iteratively merged based on similarity or distance until all points belong to a single cluster or a stopping criterion is met.

Key Steps:

1. Treat each data point as a separate cluster.

2. Compute distances between all clusters using a linkage criterion (e.g., single, complete, average).

3. Merge the two closest clusters.

4. Repeat steps 2–3 until the desired number of clusters is reached or all points form one cluster.

Agglomerative clustering is often visualized using a dendrogram, which shows the sequence of cluster merges.

#20. What makes Silhouette Score a better metric than just inertia for model evaluation?

-> While inertia (within-cluster sum of squared distances) measures how compact clusters are, it only considers distance to centroids and always decreases with more clusters, making it hard to compare different models or determine the optimal number of clusters reliably.

###Silhouette Score, on the other hand, considers both:

1. Cohesion – how close a point is to other points in its own cluster.

2. Separation – how far a point is from points in the nearest neighboring cluster.

A higher Silhouette Score indicates clusters that are well-separated and compact, providing a more holistic assessment of clustering quality. Unlike inertia, it works even when clusters are unequal in size or density and is not biased by the number of clusters.

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, random_state=42)

# Range of k values
k_values = range(2, 6)
silhouette_scores = []

# Compute silhouette scores
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Plot silhouette scores as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

```python
#31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
#  decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters
X, _ = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=2.0,   # larger std → more overlap
    random_state=42
)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Create a mesh grid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

# Predict cluster for each point in the grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    marker='x'
)
plt.title("K-Means Clustering with Decision Boundaries (Overlapping Data)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
#34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
#    line plot

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate inertia for different K values
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot inertia values
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for K-Means (Breast Cancer Dataset)")
plt.show()
```

```python
#40.Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D
#   scatter plot.

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load the Digits dataset
digits = load_digits()
X = digits.data

# Reduce to 3 principal components
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X)

# Apply K-Means clustering (10 clusters for digits 0-9)
kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X_pca)

# 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],
    c=labels, cmap='tab10', s=50
)
ax.set_title("K-Means Clustering on Digits Dataset (3D PCA)")
ax.set_xlabel("PC 1")
ax.set_ylabel("PC 2")
ax.set_zlabel("PC 3")
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```

```python
#42. Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering.
#   Visualize in 2D

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce to 2 dimensions using PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Apply Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = agglo.fit_predict(X_pca)

# Visualize clusters in 2D
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='coolwarm')
plt.title("Agglomerative Clustering on Breast Cancer Dataset (PCA Reduced)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
#43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
#    side-by-side

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN

# Generate noisy circular data
X, _ = make_circles(n_samples=400, noise=0.05, factor=0.5, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kmeans.fit_predict(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Plot side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# K-Means result
axes[0].scatter(X[:, 0], X[:, 1], c=labels_kmeans, cmap='coolwarm')
axes[0].scatter(kmeans.cluster_centers_[:, 0],
                kmeans.cluster_centers_[:, 1],
                marker='X', s=150, c='black')
axes[0].set_title("K-Means Clustering")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# DBSCAN result
axes[1].scatter(X[:, 0], X[:, 1], c=labels_dbscan, cmap='coolwarm')
axes[1].set_title("DBSCAN Clustering")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

plt.suptitle("Clustering Noisy Circular Data: K-Means vs DBSCAN")
plt.show()
```

```python
#44.Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply K-Means clustering (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Compute silhouette scores for each sample
silhouette_vals = silhouette_samples(X, labels)

# Sort silhouette values for plotting
y_ticks = np.arange(len(silhouette_vals))
sorted_silhouette_vals = np.sort(silhouette_vals)

# Plot silhouette coefficients
plt.figure(figsize=(8, 6))
plt.barh(y_ticks, sorted_silhouette_vals, color='skyblue')
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")
plt.title("Silhouette Coefficient for Each Sample (Iris Dataset, K-Means)")
plt.show()

# Optional: overall average silhouette score
avg_score = silhouette_score(X, labels)
print(f"Average Silhouette Score: {avg_score:.3f}")
```

```python
#46.Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
#  features)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

# Load the Wine dataset
wine = load_wine()
X = wine.data
feature_names = wine.feature_names

# Create a DataFrame for first 4 features
df = pd.DataFrame(X[:, :4], columns=feature_names[:4])

# Apply K-Means clustering (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X[:, :4])

# Visualize clusters with pairplot
sns.pairplot(df, hue='Cluster', palette='tab10')
plt.suptitle("K-Means Cluster Assignments (Wine Dataset, First 4 Features)", y=1.02)
plt.show()
```

```python
#47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
#    count

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate synthetic noisy blobs
X, _ = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=[0.5, 1.0, 2.0],  # varying densities for noise
    random_state=42
)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

# Count number of noise points and clustered points
num_noise = np.sum(labels == -1)
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
num_clustered_points = len(labels) - num_noise

print("Number of clusters identified:", num_clusters)
print("Number of noise points:", num_noise)
print("Number of points assigned to clusters:", num_clustered_points)
```
